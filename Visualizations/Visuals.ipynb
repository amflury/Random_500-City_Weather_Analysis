```python
# Dependencies
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
cities = pd.read_csv("../Resources/cities.csv")
cities.head()
```

```python
cities_df = pd.DataFrame(cities)
cities_df
```

```python
#Define sections for comparison
latitude = cities['Lat']
max_temp = cities['Max Temp']
humidity =  cities['Humidity']
cloudiness = cities['Cloudiness']
wind_speed = cities['Wind Speed']
```

```python
#Lat vs Max Tempt
plt.scatter(latitude, max_temp, color='b')

plt.title("City Latitude vs. Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature(F)")
plt.grid()
plt.xlim(-60, 80)
plt.ylim(0, 120)
plt.style.use('seaborn')
plt.grid(color='white')
```

```python
#Lat vs Humidity
plt.scatter(latitude, humidity, color='b')

plt.title("City Latitude vs. Humidity")
plt.xlabel("Latitude")
plt.ylabel("Humidity")
plt.grid()
plt.xlim(-60, 80)
plt.ylim(0, 120)
plt.style.use('seaborn')
plt.grid(color='white')
```

```python
#Lat vs Cloudiness
plt.scatter(latitude, cloudiness, color='b')

plt.title("City Latitude vs. Cloudiness")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness")
plt.grid()
plt.xlim(-60, 80)
plt.ylim(-20, 120)
plt.style.use('seaborn')
plt.grid(color='white')
```

```python
#Lat vs Wind Speed
plt.scatter(latitude, wind_speed, color='b')

plt.title("City Latitude vs. Wind Speed")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed")
plt.grid()
plt.xlim(-60, 80)
plt.ylim(0, 35)
plt.style.use('seaborn')
plt.grid(color='white')
```
